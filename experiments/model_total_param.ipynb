```python
import tensorflow as tf
```

```python
path = 'artifacts/checkpoints/model/'
checkpoints_name = 'exp'
checkpoints_extension = '.keras'
checkpoints_num = 5
```

```python
for i in range(1, checkpoints_num+1):
    _model_path = path + checkpoints_name + str(i) + checkpoints_extension
    print(f'####### {_model_path} #######')
    model = tf.keras.models.load_model(_model_path)
    model.summary()
    tf.keras.backend.clear_session()
```
